## Percentage Metrics

```python
#| export
from typing import Union
ArrayLike = np.ndarray|list|pd.Series

def MAPE(
    y_true: ArrayLike, # Actual values.
    y_pred: ArrayLike # Predicted values.
) -> float: # Mean absolute percentage error.
    "Calculate mean absolute percentage error."
    if len(y_true) != len(y_pred):
        raise ValueError("Input arrays must have the same length.")
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    return round(np.mean(np.abs((y_true - y_pred) / y_true)), 2)
```

```python
#| echo: false
docments(MAPE, full=True)
DocmentTbl(MAPE)
```

```python
y_true = [100, 200, 300]
y_pred = [90, 210, 330]
test_eq(MAPE(y_true, y_pred), 0.08)
test_fail(lambda: MAPE([1, 2], [1]), contains="same length")
```

```python
#| export

ArrayLike = np.ndarray|list|pd.Series
def SMAPE(
    y_true:ArrayLike, # actual values
    y_pred:ArrayLike,  # predicted values
) -> float: # symmetric mean absolute percentage error in percent
    "Calculate symmetric mean absolute percentage error."
    if len(y_true) != len(y_pred):
        raise ValueError("Input arrays must have the same length.")
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    return 1 / len(y_true) * np.sum(
        2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)) * 100
    )
```

```python
#| echo: false
DocmentTbl(SMAPE) 
```

```python
test_eq(round(SMAPE(y_true, y_pred), 2), 8.31)
test_fail(lambda: SMAPE([1, 2], [1]), contains="same length")
```

```python
#| export
def WMAPE(
    y_true: np.ndarray|list|pd.Series, # actual values
    y_pred: np.ndarray|list|pd.Series  # predicted values
) -> float: # weighted mean absolute percentage error
    "Calculate weighted mean absolute percentage error."
    if len(y_true) != len(y_pred):
        raise ValueError("Input arrays must have the same length.")
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    return np.sum(np.abs(y_true - y_pred)) / np.sum(y_true)
```

```python
test_eq(round(WMAPE(y_true, y_pred), 3), 0.083)
test_fail(lambda: WMAPE([1, 2], [1]), contains="same length")
```

## Absolute and Squared Error Metrics

```python
#| export
def MAE(
    y_true: np.ndarray|list|pd.Series, # actual values
    y_pred: np.ndarray|list|pd.Series  # predicted values
) -> float: # mean absolute error
    "Calculate mean absolute error."
    if len(y_true) != len(y_pred):
        raise ValueError("Input arrays must have the same length.")
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    return np.mean(np.abs(y_true - y_pred))
```

```python
#| export
def MSE(
    y_true: np.ndarray|list|pd.Series, # actual values
    y_pred: np.ndarray|list|pd.Series  # predicted values
) -> float: # mean squared error
    "Calculate mean squared error."
    if len(y_true) != len(y_pred):
        raise ValueError("Input arrays must have the same length.")
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    return np.mean((y_true - y_pred) ** 2)
```

```python
test_eq(round(MSE(y_true, y_pred), 2), 366.67)
test_fail(lambda: MSE([1, 2], [1]), contains="same length")
```

```python
#| export
def RMSE(y_true:ArrayLike,
         y_pred:ArrayLike) -> float:
    
    """
    Calculate Root Mean Square Error (RMSE).

    Parameters:

    - y_true: Array of true values.
    - y_pred: Array of predicted values.

    Returns:

    - rmse: Root Mean Square Error.
    """
    # Ensure both arrays have the same length
    if len(y_true) != len(y_pred):
        raise ValueError("Input arrays must have the same length.")
    # Convert to numpy arrays for element-wise operations
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    return np.sqrt(np.mean((y_true - y_pred) ** 2))
```
